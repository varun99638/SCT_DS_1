## 6. Data Visualization

### Histogram of Population Distribution (2025)

The histogram below illustrates the distribution of population values for all entries in the dataset for the year 2025. It helps us understand how population is distributed across different countries and aggregate entities.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

plt.hist(df["2025"], bins=30, edgecolor="black")

plt.title("Distribution of Population in 2025")
plt.xlabel("Population")
plt.ylabel("Number of Entries")

plt.grid(True, linestyle="--", alpha=0.6)

plt.show()
```

### Histogram with Logarithmic Scale### Top 10 Most Populous Entries (2025)

The following bar chart displays the ten entries with the highest population in the year 2025.
Since population values vary from thousands to billions, a logarithmic scale is used on the x-axis to better visualize the distribution.

```python
top10 = df[["Country Name", "2025"]].sort_values(
    by="2025", ascending=False
).head(10)

plt.figure(figsize=(12, 6))

plt.barh(top10["Country Name"], top10["2025"])

plt.title("Top 10 Most Populous Entries (2025)")
plt.xlabel("Population")
plt.ylabel("Country / Region")

plt.gca().invert_yaxis()

plt.grid(axis="x", linestyle="--", alpha=0.6)

plt.show()
```
